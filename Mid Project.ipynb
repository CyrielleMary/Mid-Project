## Hypothesis 

I want to know if the full data shows the same trend as my biathlon places

To do so, I will do 2 hypothesis

=> does the season starts earlier than the mean of my sample?

=> does the season ends later than the mean of my sample?



I want also to know if the density of snow is higher in the places where there is biathlon world cup.

## HYPOTHESIS 1

```python
# I need to get rid of the years and places where there is no snow
    # from the full data
    # from the sample
```

```python
data2 = data1[(data1.beginning_of_season != 0)]
```

```python
data2.shape
```

```python
data2['beginning_of_season'].min()
```

```python
snowfall_wc_biathlon2 = snowfall_wc_biathlon[(snowfall_wc_biathlon.beginning_of_season !=0)]
```

```python
snowfall_wc_biathlon2.beginning_of_season.mean()
```

```python
#Mean Sample
#beginning_of_season = 163
```

```python
import scipy.stats
from scipy.stats import ttest_1samp
```

```python
# H0 = The beginning of the season <= 163
# H1 = The beginning of the season > 163
```

```python
stat, pval = ttest_1samp(data2['beginning_of_season'], 163)
```

```python
print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval/2)
```

```python
#The pvalue is 0 and the stat is negative, as the sign < from our H0, we can accept the H0
```

```python
data2['beginning_of_season'].mean()
```
